# Cuisine analysis

```python
df_swiggy["Cuisine"].unique()
```

```python
cuisine_dict = {}
for i in df_swiggy['Cuisine'].unique():
    new_Cuisines = i.split(',')
    for second_Cuisine in new_Cuisines:
        Cuisine = second_Cuisine.lstrip(' ')
        if Cuisine in cuisine_dict:
            cuisine_dict[Cuisine] = cuisine_dict[Cuisine] + 1
        else: 
            cuisine_dict[Cuisine] = 1
        
print(cuisine_dict)
```
